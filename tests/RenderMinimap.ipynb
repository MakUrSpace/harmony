```python
        camImage = hm.buildMiniMap(objectsAndColors=hm.objectsAndColors)
        bounds = hm.cc.realSpaceBoundingBox()
        print(bounds)
            
        x, y, w, h = bounds[:2]
        x, y, w, h = int(x), int(y), int(w), int(h)
        
        shift_x = 0
        shift_y = 0
        if x < 0: shift_x = -x
        if y < 0: shift_y = -y

        map_x = x + shift_x
        map_y = y + shift_y
        
        margin = 150
        
        crop_x = max(0, map_x - margin)
        crop_y = max(0, map_y - margin)
        
        crop_w = w + margin * 2
        crop_h = h + margin * 2
        
        if crop_w > 0 and crop_h > 0:
            img_h, img_w = camImage.shape[:2]
            end_x = min(img_w, crop_x + crop_w)
            end_y = min(img_h, crop_y + crop_h)
            print(f"Cropping Minimap to: `Y:[{crop_y}:{end_y}], X:[{crop_x}:{end_x}]`")
```

```python
import harmony
from harmony.HarmonyMachine import HexCaptureConfiguration, HarmonyMachine
from harmony.harmonyServer import render_minimap
from matplotlib import pyplot as plt
from time import monotonic
```

```python
cc = HexCaptureConfiguration()
hm = HarmonyMachine(cc=cc)
hm.cycle()
plt.imshow(hm.getCameraImagesWithChanges()['0'])
```

```python
start = monotonic()
m = hm.buildMiniMap()
print(monotonic() - start)
plt.imshow(m)
```

```python
start = monotonic()
m = render_minimap(hm, encode=False)
print(monotonic() - start)
plt.imshow(m)
```
